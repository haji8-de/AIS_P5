```python
# 필요한 라이브러리를 먼저 불러옵니다.
import pandas as pd
import requests
```

```python
# 반복문을 위한 변수를 먼저 지정해줍니다.
paged = 1
store_list = []

# json을 반복적으로 불러 올 수 있게 반복문을 이용합니다.
while True:
    # 개발자 도구를 이용하여 찾은 url
    url = f"https://theborndb.theborn.co.kr/wp-json/api/get_store/?state=9&category=275&depth1=&depth2=&paged={paged}&search_string="

    # requests를 이용하여 json을 불러옵니다.
    response = requests.get(url)
    temp = response.json()['results'] # 가게 정보는 json의 results에 들어있습니다.

    # 정보가 끝난 페이지에서는 results가 None으로 나와 break해줍니다.
    if temp is None:
        break

    # page마다 나오는 results 값들을 list에 합쳐줍니다.
    store_list = store_list + temp

    # 페이지 수를 하나씩 높혀주어 계속해서 json을 불러올 수 있게합니다.
    paged = paged + 1
```

```python
# DataFrame으로 변환합니다.
paiks = pd.DataFrame(store_list)
paiks
```
