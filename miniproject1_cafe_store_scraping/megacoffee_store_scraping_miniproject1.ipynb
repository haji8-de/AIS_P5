```python
# 필요한 라이브러리를 먼저 불러옵니다.
import pandas as pd
import requests
from bs4 import BeautifulSoup as bs     
```

```python
# 개발자 도구를 이용하여 찾은 url
url = 'https://www.mega-mgccoffee.com/store/find/store_search.php?store_search=+'
response = requests.get(url)
soup = bs(response.text, 'html.parser')
```

```python
# 컬럼들을 위해 list를 생성합니다.
store_list = []
num_list = []
address_list = []
addr_list = []

tag_b = soup.select('b') # soup를 확인해보니 <b>에 매장명이 있어 따로 추출
# 매장명 컬럼을 text로 추출하고 리스트에 넣습니다.
for i in tag_b:
    store_list.append(i.text)

# 나머지 데이터의 태그 위치 확인 및 추출
tag_div = soup.select('div.cont_text_inner.cont_text_info')
temp_list = []

# tag_div의 데이터에 ' '가 너무 많아 split()으로 나누어 리스트에 저장합니다.
for i in tag_div:
    temp_list.append(i.text.split())

# num_list에 전화번호를, address_list에 주소를 넣는 코드    
for index in range(len(temp_list)):
    address_list.append(temp_list[index][:-1])
    num_list.append(temp_list[index][-1])

# address_list를 정제합니다.
for index in range(len(address_list)):
    addr_list.append(' '.join(address_list[index]))
```

```python
# mega dataframe 데이터 만들기
mega = pd.DataFrame()
mega['매장명'] = store_list
mega['전화번호'] = num_list
mega['주소'] = addr_list
```
