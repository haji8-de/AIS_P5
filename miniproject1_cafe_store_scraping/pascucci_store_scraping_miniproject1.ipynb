```python
# 필요한 라이브러리를 먼저 불러옵니다.
import pandas as pd
```

```python
def get_one_page_store(page):
    url= f"https://www.caffe-pascucci.co.kr/store/storeList.asp?page={page}"
    table=pd.read_html(url)

    #매장 정보가 들어있는 리스트 인덱스를 데이터 프레임에 저장합니다.
    df=table[0]
            
    return df
```

```python
# 반복문을 위한 변수를 먼저 지정해줍니다.
pascucci_store = []
page = 1

# 페이지를 반복적으로 불러 올 수 있게 반복문을 이용합니다.
while True:

    temp = get_one_page_store(page)

    # page마다 나오는 results 값들을 list에 합쳐줍니다.
    pascucci_store.append(temp)

    # 마지막 매장의 번호는 1이라서 break해줍니다.
    if temp.iloc[-1]["번호"] == 1:
        break

    # 페이지 수를 하나씩 높혀주어 계속해서 페이지를 불러올 수 있게합니다.
    page += 1
```

```python
# 데이터 프레임을 병합해줍니다.
pascucci = pd.concat(pascucci_store)
pascucci
```
