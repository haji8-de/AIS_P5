```python
# 필요한 라이브러리를 먼저 불러옵니다.
import pandas as pd
import requests
```

```python
def storelist(pageNo):
    url = f"https://www.hollys.co.kr/store/korea/korStore2.do?pageNo={pageNo}&sido=&gugun=&store="
    
    # requests를 이용하여 정보를 불러옵니다
    response = requests.get(url)
    df_table = pd.read_html(response.text)[0]
    return df_table
```

```python
# 반복문을 위한 변수를 먼저 지정해줍니다.
pageNo = 1
hollys = pd.DataFrame()

# 페이지를 반복적으로 불러 올 수 있게 반복문을 이용합니다.
while True:

    # 정보가 끝난 페이지에서는 지역 칼럼에 '등록된 지점이 없습니다' 라고 나와 break해줍니다.
    if storelist(pageNo).iloc[0]['지역'] == '등록된 지점이 없습니다.':
        break

    # page마다 나오는 results 값들을 temp에 저장시키고 df에 합쳐줍니다.
    temp =  storelist(pageNo)
    hollys = hollys.concat([hollys, temp])

    # 페이지 수를 하나씩 높혀주어 계속해서 불러올 수 있게합니다.
    pageNo = pageNo + 1
```

```python
# 데이터 프레임의 인덱스를 정리해줍니다.
hollys = hollys.reset_index(drop=True)
hollys
```
